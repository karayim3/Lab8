```python
import numpy as np
import matplotlib.pyplot as plt
import math
from google.colab import files

uploaded = files.upload()

for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))
```

```python
time, run1, run2, run3, run4, run5 = np.loadtxt('angularvelocity20g.txt', skiprows=1, unpack=True)

ax2 = plt.axes([0, 0, 2, 1.5])
plt.plot(time, abs(run1), color='red', label='Run 1')  # the velocity is negative so taking the absolute value will make it easier to compare the graphs
a, b = np.polyfit(time, abs(run1), 1)  # a is the angular acceleration (slope of angular velocity)
plt.plot(time, a*time+b, '--', color='red', label='Run 1 Best Fit')
uncertainty1 = []
for number in range(len(time)):
  real = abs(run1[number])
  best = a * time[number]
  uncertainty1.append(abs(real-best))
print("Angular acceleration in Run 1:", a, 'rad/s^2')
print("Uncertainty for angular acceleration (rad/s^2) in Run 1:", uncertainty1)

plt.plot(time, run2, color='orange', label='Run 2')
c, d = np.polyfit(time, run2, 1)
plt.plot(time, c*time+d, '--', color='orange', label='Run 2 Best Fit')
uncertainty2 = []
for number in range(len(time)):
  real = run2[number]
  best = a * time[number]
  uncertainty2.append(abs(real-best))
print("Angular acceleration in Run 2:", c, 'rad/s^2')
print("Uncertainty for angular acceleration (rad/s^2) in Run 2:", uncertainty2)

plt.plot(time, run3, color='yellow', label='Run 3')
e, f = np.polyfit(time, run3, 1)
plt.plot(time, e*time+f, '--', color='yellow', label='Run 3 Best Fit')
uncertainty3 = []
for number in range(len(time)):
  real = run3[number]
  best = a * time[number]
  uncertainty3.append(abs(real-best))
print("Angular acceleration in Run 3:", e, 'rad/s^2')
print("Uncertainty for angular acceleration (rad/s^2) in Run 3:", uncertainty3)

plt.plot(time, run4, color='green', label='Run 4')
g, h = np.polyfit(time, run4, 1)
plt.plot(time, g*time+h, '--', color='green', label='Run 4 Best Fit')
uncertainty4 = []
for number in range(len(time)):
  real = run4[number]
  best = a * time[number]
  uncertainty4.append(abs(real-best))
print("Angular acceleration in Run 4:", g, 'rad/s^2')
print("Uncertainty for angular acceleration (rad/s^2) in Run 4:", uncertainty4)

plt.plot(time, run5, color='blue', label='Run 5')
i, j = np.polyfit(time, run5, 1)
plt.plot(time, i*time+j, '--', color='blue', label='Run 5 Best Fit')
uncertainty5 = []
for number in range(len(time)):
  real = run5[number]
  best = a * time[number]
  uncertainty5.append(abs(real-best))
print("Angular acceleration in Run 5:", i, 'rad/s^2')
print("Uncertainty for angular acceleration (rad/s^2) in Run 5:", uncertainty5)

plt.legend()
plt.title("Angular Velocity (rad/s) for a 20g Mass")
plt.xlabel('Time (s)')
plt.ylabel('Angular Velocity (rad/s)')
plt.show()
```
